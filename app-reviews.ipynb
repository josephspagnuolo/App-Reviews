#### **PREPROCESSING THE TEXT**

##### Imports and Defined Functions:

```python
!pip install -qq contractions
!pip install -qq nltk
!pip install -qq num2words
import nltk
import string
import contractions
from num2words import num2words
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer
import re
import unicodedata
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
stop_words.add('game')
wnl = WordNetLemmatizer()

def sep_num_words_func(text):
    return re.sub(r"([0-9]+(\.[0-9]+)?)",r" \1 ", text).strip()

def remove_accented_chars_func(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')

def remove_extra_whitespaces_func(text): 
    return re.sub(r'^\s*|\s\s*', ' ', text).strip()

def num_to_words(text):
    after_spliting = text.split()
    for index in range(len(after_spliting)):
        if after_spliting[index].isdigit():
            after_spliting[index] = num2words(after_spliting[index])
    numbers_to_words = ' '.join(after_spliting)
    return numbers_to_words

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

##### My App - Parlera

```python
df1 = pd.read_csv('com.enjoyingfoss.parlera.csv')

# I. Remove punctuations
# II. Remove special characters and emojis

df1['Contractions'] = df1['Review'].apply(lambda x: [contractions.fix(word) for word in x.split()])
df1['Back to String'] = [' '.join(map(str, l)) for l in df1['Contractions']]
df1['Back to String'].replace('1st', 'first', regex=True, inplace=True)
df1['Back to String'].replace('2nd', 'second', regex=True, inplace=True)
df1['Back to String'].replace('3rd', 'third', regex=True, inplace=True)
df1['Separated'] = df1['Back to String'].apply(sep_num_words_func)
df1['Punctuation'] = df1['Separated'].str.replace(r'[^\w\s]+', "  ")
df1['Punctuation'] = df1['Punctuation'].apply(remove_accented_chars_func)

# III. Turn numbers into text

df1['Convert Numbers to Words'] = df1['Punctuation'].apply(num_to_words)
df1['Convert Numbers to Words'] = df1['Convert Numbers to Words'].str.replace(r'[^\w\s]+', "  ")

# IV. Remove extra white spaces

df1['Convert Numbers to Words'] = df1['Convert Numbers to Words'].apply(remove_extra_whitespaces_func)

# V. Turn all words into lowercase

df1['Lowercase'] = df1['Convert Numbers to Words'].str.lower()

# VI. Remove stop words

df1['Tokenized'] = df1['Lowercase'].apply(wordpunct_tokenize)
df1['Stopwords'] = df1['Tokenized'].apply(lambda x: [word for word in x if word not in stop_words])

# VII. Lemmatize the reviews

df1['POS Tags'] = df1['Stopwords'].apply(nltk.tag.pos_tag)
df1['Wordnet'] = df1['POS Tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
df1['Final Pre-processed Review'] = df1['Wordnet'].apply(lambda x: [wnl.lemmatize(word, tag) for word, tag in x])
df1['Final Pre-processed Review'] = df1['Final Pre-processed Review'].apply(lambda x: [word for word in x if word not in stop_words])

# VIII. Output 15 sample pre-processed reviews

df1.drop('Reviewer Name', inplace=True, axis=1)
df1.drop('Rating', inplace=True, axis=1)
dict2 = {'Review': 'Original Review'}
df1.rename(columns=dict2, inplace=True)
df1.drop('Contractions', inplace=True, axis=1)
df1.drop('Back to String', inplace=True, axis=1)
df1.drop('Separated', inplace=True, axis=1)
df1.drop('Punctuation', inplace=True, axis=1)
df1.drop('Convert Numbers to Words', inplace=True, axis=1)
df1.drop('Lowercase', inplace=True, axis=1)
df1.drop('Tokenized', inplace=True, axis=1)
df1.drop('Stopwords', inplace=True, axis=1)
df1.drop('POS Tags', inplace=True, axis=1)
df1.drop('Wordnet', inplace=True, axis=1)
df1.head(15)
```

#### **SENTIMENT ANALYSIS**

##### Imports and Defined Functions:

```python
!pip install -qq vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
analyzer = SentimentIntensityAnalyzer()

def getPolarity(text):
   return TextBlob(text).sentiment.polarity
```

```python
from collections import defaultdict
import gensim
from gensim import corpora
from gensim.corpora.dictionary import Dictionary

dictionary = Dictionary(df2['Final Pre-processed Review'])
dictionary.filter_extremes(no_below=5, no_above=0.5, keep_n=1000)
corpus = [dictionary.doc2bow(doc) for doc in df2['Final Pre-processed Review']]
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(corpus, num_topics=10, id2word = dictionary, passes=10, random_state=0, eval_every=None)
for idx, topic in ldamodel.show_topics(formatted=False, num_words= 10):
    print('Topic {} described by Words: {}\n'.format(idx+1, ', '.join([w[0] for w in topic])))
```
